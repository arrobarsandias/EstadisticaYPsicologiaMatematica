# Práctica: Regresión Múltiple

Eres parte de un grupo de investigación enfocado en el comportamiento social. Actualmente, tu grupo se encuentra analizando una base de datos de comportamiento antisocial en Estados unidos. Cuentas con las siguientes variables:

- moralIntegration: Compuesto de la tasa de delincuencia y los gastos de bienestar.
- heterogeneity: Heterogeneidad étnica, a partir de porcentajes de residentes no blancos y blancos nacidos en el extranjero.
- mobility: Movilidad geográfica a partir de porcentajes de residentes que entran y salen de la ciudad.
- region: E = Northeast, MW = Midwest, S = Southeast, W = West 

Para empezar, se te ha asignado la exploración de posibles variables que puedan influir en el comportamiento antisocial en las diferentes ciudades. 

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv(
    "https://socialsciences.mcmaster.ca/jfox/Books/Applied-Regression-3E/datasets/Angell.txt",
    sep=r"\s{1,}",
    engine="python",
)
angell_data = pd.get_dummies(df, drop_first=True)
angell_data
```

## Parte 1: Diseño

1. Estandariza los datos y explora las correlaciones entre las variables a través de un mapa de calor. ¿Qué observas?  

```python
# Transformacion Z
scaler = StandardScaler()
scaled_data = scaler.fit_transform(angell_data)
scaled_angell_data = pd.DataFrame(scaled_data, columns=angell_data.columns)
scaled_angell_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(scaled_angell_data.corr())
plt.show()
```

Podemos apreciar correlaciones moderadas (positivas y negativas) entre la variable moralIntegration y las demás variables.

2. Implementa un modelo de regresión múltiple que tome como variable dependiente a la variable `workingMom` y como variables independientes al resto.

```python
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
import statsmodels.api as sm

# Fit regression model
model = smf.ols(
    "moralIntegration ~ heterogeneity + mobility + region_MW + region_S + region_W",
    data=scaled_angell_data,
).fit()
```

## Parte 2: Verificación de supuestos

3. Verifica el supuesto de homocedasticidad. En caso el supuesto no se cumpla, aplica una medida correctiva en el modelo previamente especificado.

```python
name = ["Lagrange multiplier statistic", "p-value", "f-value", "f p-value"]
test = sms.het_breuschpagan(model.resid, model.model.exog)
list(zip(name, test))
```

La prueba de Lagrange nos permite afirmar que el supuesto de homocedasticidad se cumple.

4. Verifica el supuesto de normalidad. En caso el supuesto no se cumpla, aplica una medida correctiva en el modelo previamente especificado.

```python
import pingouin as pg

pg.qqplot(model.resid)
plt.show()
```

```python
name = ["Jarque-Bera", "Chi^2 two-tail prob.", "Skew", "Kurtosis"]
test = sms.jarque_bera(model.resid)
list(zip(name, test))
```

La prueba de Jaque-Bera nos permite afirmar que el supuesto de normalidad se cumple.

## Estimación y evaluación

6. Examina la significancia del modelo y su coeficiente de determinación. ¿Qué observas?

```python
model.summary()
```

Podemos apreciar que el modelo de regresión es estadísticamente significativo. Es decir, su coeficiente de determinación es diferente de cero. A nivel descriptivo podemos apreciar que el modelo propuesto explica el 68% de la varianza. 

7. Examina la significancia de los coeficientes de regresión. ¿Puedes simplificar el modelo?

```python
new_model = smf.ols(
    "moralIntegration ~ heterogeneity + region_MW + region_S + region_W",
    data=scaled_angell_data,
).fit()
new_model.summary()
```

El coeficiente de regresión de la variable `mobility` no es estadísticamente significativo. Podemos optar por retirar dicha variable del modelo, de modo que contamos con `heterogeneity` y `region` como predictores de `moralIntregration`.

10. Examina la multicolinealiad del modelo. ¿Qué observas? 

```python
X = pd.DataFrame(new_model.model.exog, columns=new_model.params.index)
X.corr()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# the independent variables set
X = new_model.model.exog

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = new_model.params.index

# calculating VIF for each feature
vif_data["VIF"] = [
    variance_inflation_factor(X, i)
    for i in range(new_model.model.exog.shape[1])
]
vif_data
```

Si bien existe una correlación especialmente fuerte entre las variables `heterogeneity` y `region_S`, el factor VIF se encuentra dentro de los rangos aceptables. Podemos concluir preliminarmente que el modelo no presenta problemas de multicolinealidad.
