```python
from nhanes.load import load_NHANES_data

nhanes_data = load_NHANES_data()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

minor_nhanes_data = nhanes_data.query("AgeInYearsAtScreening < 18")
sns.scatterplot(x="WeightKg", y="StandingHeightCm", data=minor_nhanes_data)
plt.xlabel("Peso (Kg)", size=14)
plt.ylabel("Altura (cm)", size=14);
```

```python
adult_nhanes_data = nhanes_data.query("AgeInYearsAtScreening > 18")
sns.boxplot(x="WalkOrBicycle", y="StandingHeightCm", data=adult_nhanes_data)
plt.xlabel("¿Activo físicamente?", size=14)
plt.xticks([0, 1], ["No", "Sí"])
plt.ylabel("Altura (cm)", size=14);
```
