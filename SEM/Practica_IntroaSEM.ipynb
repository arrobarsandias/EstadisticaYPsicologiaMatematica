```python
## Path diagram with python
import semopy

desc = '''

# Measurement
e1 =~ Intent
e2 =~ Behavior

# Structual
Intent ∼ Attitude + SubNorm + PBC 
Behavior ∼ PBC
Behavior ∼ Intent

# Covariances
SubNorm ∼∼ PBC
Attitude ∼∼ PBC
Attitude ∼∼ SubNorm

'''
g = semopy.semplot(desc, "out.png", plot_covs=True)
g
```

```python
model = semopy.Model(desc)
res = model.fit(df)
```
