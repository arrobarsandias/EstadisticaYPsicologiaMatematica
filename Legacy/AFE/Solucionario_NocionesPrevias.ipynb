```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_spss("smoke.sav")
df.dropna(inplace=True)
df
```

## Parte 1: Medidas estadísticas como vectores

Calcula la varianza de la variable s8 como una proporción de la norma.

```python
import numpy as np

# Escribe tu respuesta aquí
n = df["s8"].size
x_centered = df["s8"] - df["s8"].mean()
var = (1 / n) * np.linalg.norm(x_centered) ** 2
var
```

Calcula la correlación de las variables s7 y s8 como un producto interno normalizado:

```python
# Escribe tu respuesta aquí
x_centered = df["s7"] - df["s7"].mean()
y_centered = df["s8"] - df["s8"].mean()

cor = np.inner(x_centered, y_centered) / (
    np.linalg.norm(x_centered) * np.linalg.norm(y_centered)
)
cor
```

Opcional: Calcula la matriz de covarianza de los datos usando multiplicación de matrices. 

```python
# Escribe tu respuesta aquí
X_centered = df - df.mean(axis=0)
cov_matrix = 1 / n * (X_centered.T @ X_centered)
cov_matrix
```

## Parte 2: Análisis Factorial Común

```python
# Estandarizamos los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
df = pd.DataFrame(scaled_data, columns=df.columns.values)
df
```

Calcula la matriz de covarianza de los datos.

```python
# Escribe tu respuesta aquí
covariance_matrix = df.cov()
covariance_matrix
```

Aplica la descomposición de valores propios a la matriz de covarianza de los datos.

```python
# Escribe tu respuesta aquí
values, vectors = np.linalg.eig(covariance_matrix)
pd.DataFrame(vectors, columns=values, index=df.columns.values)
```

Calcula las cargas factoriales del modelo.

```python
# Escribe tu respuesta aquí
factor_loadings = np.sqrt(values) * vectors
loadings = pd.DataFrame(factor_loadings, index=df.columns.values)
loadings.columns += 1
loadings = loadings.add_prefix("F")
loadings
```
